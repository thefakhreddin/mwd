```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec

# ---------- load ----------
CSV_FILE = 'imu_challenge_dataset_ext.csv'

# first line is a title row, skip it
df = pd.read_csv(CSV_FILE, skiprows=1)
print(f'Rows: {len(df):,}   Columns: {list(df.columns)}')
df.head(3)
```

```python
TIME_COL = 'time_s'
data_cols = [c for c in df.columns if c != TIME_COL]

n = len(data_cols)
fig, axes = plt.subplots(n, 1, figsize=(14, 3 * n), sharex=True)
if n == 1:
    axes = [axes]

for ax, col in zip(axes, data_cols):
    ax.plot(df[TIME_COL], df[col], linewidth=0.7)
    ax.set_title(col, fontsize=10, fontweight='bold')
    ax.set_ylabel(col)
    ax.grid(True, alpha=0.3)

axes[-1].set_xlabel('Time (s)', fontsize=10)
fig.suptitle('IMU Dataset — All Channels vs Time', fontsize=13, y=1.01)
plt.tight_layout()
plt.savefig('all_columns_vs_time.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved: all_columns_vs_time.png')
```
